```python
enron_df = pd.DataFrame(enron_dataset)

# create a header variable from the first column of the dataset
header = enron_df.iloc[0]
header
```

```python
enron_df.describe()
names_df = enron_df['Unnamed: 0']
enron_df = enron_df.drop(['Unnamed: 0'], axis=1)
```

```python
## Store the target value
classes = enron_df['poi']
classes.head()
```

```python
## Drop the Class values from data
df_no_poi = enron_df.drop(['poi'],axis=1) 
df_no_poi
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
# Fit the model to the data
kmeans.fit(enron_df)
```

```python
predicted_clusters = kmeans.predict(X=enron_df)
```

```python
enron_df.shape
```

```python
import numpy as np
from sklearn.decomposition import PCA
X = np.array(df_no_poi)
pca = PCA(n_components=2)
new_X = pca.fit_transform(X)  


print(pca.explained_variance_ratio_)  

print(pca.singular_values_) 
```

```python
temp = list(zip(new_X.tolist(),predicted_clusters.tolist(),classes.tolist()))
print(temp)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x =new_X[:,0]
y =classes
z =new_X[:,1]

ax.scatter(x, y, z, c=predicted_clusters, s=25, cmap='cool')

ax.set_xlabel('PC-1')
ax.set_ylabel('POI')
ax.set_zlabel('PC-2') 

plt.show()
```
